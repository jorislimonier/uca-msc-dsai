```r
signature <- function (g, max=10) {
  clqs <- cliques(g)
  clique_lengths <- rep(0, max)
  for (i in clqs){
    len <- length(i)
    clique_lengths[len] <- clique_lengths[len] + 1
  }
  clique_lengths
  barplot(clique_lengths)
}

signature(g)
```
