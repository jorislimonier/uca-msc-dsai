```r
source("05-utils.R")

n_clusters <- 2
pi_ <- rep(1 / n_clusters, n_clusters)
mu <- c(70, 90) # or
mu <- rnorm(n_clusters, mean(X), sd(X))
sigma <- rep(var(X), n_clusters) # variance, not std


e_step(X=X, pi_=pi_, mu=mu, sigma=sigma)
```
