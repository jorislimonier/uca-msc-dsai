### 2) Show that $X$ and $Y$ are not independent


$X$ and $Y$ are independent if the following holds:

$$
\mathbb{P}(X, Y) = \mathbb{P}(X)\mathbb{P}(Y)
$$

In other words, the joint proability of $X$ and $Y$ equals the product of their respective probabilities. We verify whether this is the case.



````python
one.check_independent()
````

We compare the above ($\mathbb{P}(X)\mathbb{P}(Y)$) with the initial distribution plotted above ($\mathbb{P}(X, Y)$) and we notice that coefficients don't match. We therefore deduce that the random variables $X$ and $Y$ are not independent.

### 4) Find $\mathbb{P}(Y=10 | X < 2)$

We first compute $\mathbb{P}(Y=10 \mid X<2)$ mathematically, using Bayes' rule. Remember that:
\begin{align*}
\Omega(X) &= \left\{ 1, 2, 3 \right\} \\
\Omega(Y) &= \left\{ 5, 10, 15 \right\}
\end{align*}

\begin{align*}
\mathbb{P}(Y=10 \mid X<2)
&= \frac{\mathbb{P}(Y=10, X<2)}{\mathbb{P}(X<2)} \\
&= \frac{\mathbb{P}(Y=10, X=1)}{\mathbb{P}(X=1)} \\
&= \frac{\mathbb{P}(Y=10, X=1)}{\sum_{y \in \Omega(X)} \mathbb{P}(X=1, Y=y)} \\
&= \frac{\mathbb{P}(Y=10, X=1)}{\mathbb{P}(X=1, Y=5) + \mathbb{P}(X=1, Y=10) + \mathbb{P}(X=1, Y=15)} \\
&= \frac{0.1}{0 + 0.1 + 0.1} \\
&= \frac{1}{2} \\
&\approx 0.5
\end{align*}

Here is the computation with code:

### 1) Approximate this function using the Laplace approximation.

The Laplace method consists in approximating a given function using a Gaussian distribution. This method makes use of the Taylor expansion of the (log-)function to approximate _f_. Let us define a function _g_ such that:

$$f(x) = \exp(\log(f(\theta)) = \exp(g(\theta))$$

Let $\theta_0$ be a maximum shared by $g$ and $f$. We use the Taylor expansion up to the second order to approximate _g_ and deduce _f_ for any point $\theta$ sufficiently close to $\theta_0$ such that:

\begin{align*}
g(\theta)
&= g(\theta_0) + g'(\theta_0)(\theta-\theta_0) + \frac{1}{2}g''(\theta_0)(\theta-\theta_0)^2 + O((\theta - \theta_0)^3)\\
g'(\theta_0) &= 0 \\
\end{align*}

We approximate $g$ as follows:

\begin{align*}
g(\theta) &\approx g(\theta_0) + \frac{1}{2}g''(\theta_0)(\theta-\theta_0)^2
\end{align*}

and we deduce $f$:
\begin{align*}
&
g(\theta) \approx g(\theta_0) + \frac{1}{2}g''(\theta_0)(\theta-\theta_0)^2 \\
\implies & 
f(\theta) \approx f(\theta_0) \exp\left[\frac{1}{2}g''(\theta_0)(\theta-\theta_0)^2\right]
\end{align*}
We recognize that the last term is the PDF of a Normal distribution:
$$
\mathcal{N}\left(\theta_0, \frac{-1}{g''(\theta_0)}\right)
$$
We find the maximum of the function by computing the derivative with respect to $\theta$ and equating to 0.



We compute the derivative:

Using $f(x) = exp(x^2 *sin(x))$ as stated above, we have:

\begin{align*}
g(\theta) & = \log(f(\theta)) = \theta^2 * \sin(\theta)\\
\frac{\delta g}{\delta \theta} (\theta) &= 2\theta \sin(\theta) + \theta^2 * \cos(\theta)
\end{align*}

therefore, using a numerical solver (Wolfram Alpha), we get that the solutions rounded to the one digit after the decimal point are $\theta_0 \in \left\{\pm 2.3, \pm 5.1, \pm 8.1, \pm 11.2 \right\}$. We want solutions $\theta_0 \in [0, 3]$ so we end up with:
$$
\mu := \theta_0 \approx 2.3
$$

Now we get the variance by computing the second order derivative of $g$ respect to $\theta$:

\begin{align*}
\frac{\delta^2 g}{\delta \theta^2} (\theta)
&= \frac{\delta}{\delta \theta} 2\theta \sin(\theta) + \theta^2 * \cos(\theta) \\
&= (2 \sin(\theta) + 2 \theta \cos(\theta)) + (2 \theta \cos(\theta) - \theta^2 \sin(\theta)) \\
&=  4 \cos(\theta) + (2 - \theta^2) \sin(\theta) \\
\end{align*}

Injecting $\theta_0$, we obtain:

$$
g''(\theta_0) = 4 \cos(\theta_0) + (2 - \theta_0^2) \sin(\theta_0)
$$

Thus:

$$
\sigma^2 := \frac{-1}{g''(\theta_0)} = \frac{-1}{4 \cos(\theta_0) + (2 - \theta_0^2) \sin(\theta_0)}
$$

Finally, the approximation will be given by:

$$
\mathcal{N}\left( \mu, \sigma^2 \right)
$$
