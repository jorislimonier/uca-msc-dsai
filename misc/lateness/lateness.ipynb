```python
import pandas as pd
import plotly.express as px

```

```python
LATENESS_PATH = "~/Documents/uca-msc-dsai/misc/lateness/"
data = pd.read_excel(LATENESS_PATH + "lateness.ods")
data.tail()

```

```python
data["start_time"] = abs(data["start_time"])
```

```python
av_late = data.groupby("teacher").mean().sort_values(by="start_time", ascending=False)
px.bar(
    data_frame=av_late,
    y="start_time",
    color="start_time",
    color_continuous_scale="Bluered",
    barmode="group",
    title="Mean lateness",
)

```

```python
sum_late = data.groupby("teacher").sum().sort_values("start_time", ascending=False)
px.bar(
    sum_late,
    y="start_time",
    color="start_time",
    color_continuous_scale="Bluered",
    barmode="group",
    title="Cumulated lateness",
)
```

```python
median_late = (
    data.groupby("teacher").median().sort_values("start_time", ascending=False)
)
px.bar(
    av_late,
    y="start_time",
    color="start_time",
    color_continuous_scale="Bluered",
    barmode="group",
    title="Median lateness",
)

```

```python
px.bar(
    data.sort_values(
        "start_time",
        ascending=False,
    ),
    x="teacher",
    y="start_time",
    color="start_time",
    color_continuous_scale="Bluered",
    barmode="group",
)
```
